```python
import pandas as pd
from tabulate import tabulate
from colorama import Fore, Style, init

init(autoreset=True)

# ===============================
# 1. Load Excel file (ALL sheets)
# ===============================
file_path = r"C:\Users\allad\Desktop\job_offer\AGS\AMRESCO_Automation\Cleaning agent\Aligned Energy 2025 Monthly Production.xlsx"

sheets = pd.read_excel(file_path, sheet_name=None)

print(f"\n📘 Loaded {len(sheets)} sheets from Excel file")

# ===============================
# 2. Collect Missing / Zero issues
#    (Skip first 3 rows)
# ===============================
issues = []

for sheet_name, df in sheets.items():

    data_df = df.iloc[3:]   # ⬅️ skip top 3 rows

    for row_idx in data_df.index:
        for col in data_df.columns:
            value = data_df.at[row_idx, col]

            # Missing values
            if pd.isna(value):
                issues.append([
                    sheet_name,
                    row_idx,
                    col,
                    "MISSING (NaN)",
                    value
                ])

            # Zero values (robust detection)
            else:
                numeric_value = pd.to_numeric(value, errors="coerce")
                if numeric_value == 0:
                    issues.append([
                        sheet_name,
                        row_idx,
                        col,
                        "ZERO",
                        value
                    ])

# ===============================
# 3. Build Issues Table
# ===============================
issues_df = pd.DataFrame(
    issues,
    columns=["Sheet", "Row_Index", "Column", "Issue_Type", "Value"]
)

# ===============================
# 4. Colored Console Output
# ===============================
if issues_df.empty:
    print(Fore.GREEN + "\n✅ No missing or zero values found across all sheets.")
else:
    def color_issue(issue):
        if issue == "MISSING (NaN)":
            return Fore.RED + issue + Style.RESET_ALL
        return Fore.YELLOW + issue + Style.RESET_ALL

    issues_df["Issue_Type"] = issues_df["Issue_Type"].apply(color_issue)

    print(Fore.CYAN + "\n🚨 DATA QUALITY ISSUES (Missing & Zero Values)\n")

    print(
        tabulate(
            issues_df,
            headers="keys",
            tablefmt="fancy_grid",
            showindex=False
        )
    )

    print(
        Fore.MAGENTA
        + f"\n📊 Total Issues Found: {len(issues_df)}"
    )
```
