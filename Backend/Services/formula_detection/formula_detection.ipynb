```python
# Print Formulas Grouped by Sheet and Column (Boxed Table Output)

import openpyxl
import pandas as pd
from tabulate import tabulate

file_path = "/content/American Polyfilm 2025 Monthly Production.xlsx"

wb = openpyxl.load_workbook(file_path, data_only=False)

data = []

for sheet_name in wb.sheetnames:
    sheet = wb[sheet_name]

    for row in sheet.iter_rows():
        for cell in row:
            if isinstance(cell.value, str) and cell.value.startswith("="):
                data.append([sheet_name, cell.column_letter, cell.value])

# Create DataFrame
df = pd.DataFrame(data, columns=["Sheet", "Column", "Formula"])

# Remove duplicate formulas within same sheet & column (clean view)
df_clean = df.drop_duplicates().sort_values(by=["Sheet", "Column"]).reset_index(drop=True)

print("\nFORMULAS BY SHEET AND COLUMN\n")
print(tabulate(df_clean, headers="keys", tablefmt="grid", showindex=False))
```
